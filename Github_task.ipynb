```python
repository_href
```

```python
repository_url_list=[]
for i in repository_href:
    repository_url_list.append("https://github.com"+i)
print(repository_url_list)
```

```python
for j in repository_url_list:
    response_html=requests.get(url=j)
    repository_name_html = bs(response_html.content, "html.parser")
    repository_name_about = repository_name_html.find('div', {'class':'Layout-sidebar'})
    repository_name_sub = repository_name_about.find_all('div', {'class':'mt-2'})
    
    repository_name_star = repository_name_sub[0].find('strong')
    repository_name_star_text=repository_name_sub[0].find('a')
    repository_name_star_textname=repository_name_star_text.text.strip().split()
    repository_name_star_no=repository_name_star.text.strip()
    
    repository_name_watch = repository_name_sub[1].find('strong')
    repository_name_watch_text=repository_name_sub[1].find('a')
    repository_name_watch_textname=repository_name_watch_text.text.strip().split()
    repository_name_watch_no=repository_name_watch.text.strip()
    
    repository_name_forks = repository_name_sub[2].find('strong')
    repository_name_forks_text=repository_name_sub[2].find('a')
    repository_name_forks_textname=repository_name_forks_text.text.strip().split()
    repository_name_forks_no=repository_name_forks.text.strip()
    
    print("{}:{},{}:{},{}:{}".format(repository_name_star_textname[1],repository_name_star_no,repository_name_watch_textname[1],repository_name_watch_no,repository_name_forks_textname[1],repository_name_forks_no),end="\n\n")
```
